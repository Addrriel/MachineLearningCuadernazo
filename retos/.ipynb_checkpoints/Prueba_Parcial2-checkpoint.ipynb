## matplotlib
- Crear una lista de años desde 1995 hasta 2015 (eje x)

- Generar 3 listas de datos aleatorios para el eje y: t0, t1, t2

- Representar las 3 secuencias en una sola figura

- Añadir una leyenda

- Nombres de las secuencias:

  - “Quito”

  - “Bogotá”

  - “Lima”

- Representar los mismos datos en 3 subgráficos (subplots) dentro de la misma figura

- Añadir:

  - Título general: “Temperaturas promedio”

  - Nombre eje x: “Año”

  - Nombre eje y: “Temperatura (°C)”

  - Leyenda en cada subplot

  - Añadir una anotación indicando el valor máximo en uno de los gráficos


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#1) Crear una lista de años desde 1995 hasta 2015 (eje x)
lista_anios = np.arange(1995, 2016)

#2) Generar 3 listas de datos aleatorios para el eje y: t0, t1, t2
t0 = np.random.randint(2100, size=(21))   
t1 = np.random.randint(2100, size=(21)) 
t2 = np.random.randint(2100, size=(21))

ciudades = ["Quito","Bogotá","Lima"]
#3) Representar las 3 secuencias en una sola figura
plt.figure()

#plt.plot(x, y, style, label, linewidth=1, alpha=1)
plt.plot(lista_anios, t0, 'co:', label=ciudades[0])
plt.plot(lista_anios, t1, 'go:' , label=ciudades[1])
plt.plot(lista_anios, t2, 'ro:',  label=ciudades[2])

plt.title("Datos aleatorios")
plt.xlabel("valores")
plt.ylabel("año")

#4) Añadir leyenda
plt.legend(loc = "best")

plt.show()
```

```python
#5) Representar los mismos datos en 3 gráficos

figura, ejes = plt.subplots(3,1)
ejes[0].plot(lista_anios, t0, 'co:', label = ciudades[0])
ejes[1].plot(lista_anios, t1, 'go:', label = ciudades[1])
ejes[2].plot(lista_anios, t2, 'ro:', label = ciudades[2])

# Título general: “Temperaturas promedio”

# Nombre eje x: “Año”
ejes[2].set_xlabel("Año")  # solo el de abajo

# Nombre eje y: “Temperatura (°C)”
ejes[0].set_ylabel("Temperatura (°C)")
ejes[1].set_ylabel("Temperatura (°C)")
ejes[2].set_ylabel("Temperatura (°C)")



# Leyenda en cada subplot
for ax in ejes:
    ax.legend(loc="best")


# Añadir una anotación indicando el valor máximo en uno de los gráficos
max_y = t0.max()
max_x = lista_anios[t0.argmax()]

max_y1 = t1.max()
max_x1 = lista_anios[t1.argmax()]

max_y2 = t2.max()
max_x2 = lista_anios[t2.argmax()]

# Añadir anotación al primer subplot
ejes[0].annotate("Valor máximo", xy=(max_x, max_y), xytext=(max_x, max_y ), arrowprops=dict(facecolor= 'red',  shrink=1.0))
ejes[1].annotate("Valor máximo", xy=(max_x1, max_y1), xytext=(max_x1, max_y1), arrowprops=dict(arrowstyle="->"))
ejes[2].annotate("Valor máximo", xy=(max_x2, max_y2), xytext=(max_x2, max_y2 ), arrowprops=dict(arrowstyle="->"))

figura.suptitle("Temperaturas promedio")

for ax in ejes:
    ax.set_ylim(0, 2200)
    ax.set_xticks(lista_anios[::5])




figura, ejes = plt.subplots(3, 1)
plt.tight_layout()
plt.show()

#anotaciones
#axes.text(x,y,’texto’)
#axes.annotate(texto, xy, xytext, xycoords, arrowprops)
```
