<h2>Reto 3 ‚Äî Funciones</h2>

<p><strong>EJERCICIO:</strong></p>

<ul>
  <li>Crear una funci√≥n que reciba dos cadenas de texto y retorne un n√∫mero.</li>
  <li>La funci√≥n imprime los n√∫meros del 1 al 100 teniendo en cuenta que:
    <ul>
      <li>Si el n√∫mero es m√∫ltiplo de 3, mostrar la primera cadena.</li>
      <li>Si el n√∫mero es m√∫ltiplo de 5, mostrar la segunda cadena.</li>
      <li>Si es m√∫ltiplo de 3 y 5, mostrar ambas cadenas concatenadas.</li>
      <li>La funci√≥n debe retornar cu√°ntas veces se imprimi√≥ el n√∫mero en lugar del texto.</li>
    </ul>
  </li>
</ul>

<p>
Presta especial atenci√≥n a la sintaxis que debes utilizar en cada uno de los casos.
Cada lenguaje sigue sus propias convenciones y debes respetarlas para que el c√≥digo se entienda.
</p>


```python
def ejercicio1 (cadena1, cadena2):
    contador = 0

    for numero in range(1,101):
        if numero % 3 == 0 and numero % 5 == 0:
            print(cadena1 + cadena2)
        elif numero % 3 == 0:
            print(cadena1)
        elif numero % 5 == 0:
            print(cadena2)
        else:
            print(numero)
            contador += 1
    return contador

cadena1 = "Soy M√∫ltiplo de 3"
cadena2 = "Soy M√∫ltiplo de 5"

resultado = ejercicio1(cadena1, cadena2)
print(resultado)
```
