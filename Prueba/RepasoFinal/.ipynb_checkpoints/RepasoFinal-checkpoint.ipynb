```python
# Escribe un programa que obtenga la mediana de una lista de números.
#Recuerda que la mediana M de una lista de números L es el número que cumple la siguiente propiedad: la mitad de los números de L son superiores a M y la otra mitad son inferiores.
#Cuando el número de elementos de L es par, se puede considerar que hay dos medianas. No obstante, en este ejercicio consideraremos que únicamente existe una mediana.

#LÓGICA
#1) Crear una lista de números
#2) Ordeno los números de la lista
#3) Creo una variable para guardar la cantidad que hay en la lista
#4) Creo un bucle if donde digo que si es impar, entonces hará la operación que sería variable // 2
#5) Si es par, entonces creamos dos variables porque habrá dos medianas. Luego las sumamos y las dividimos para dos
#6) Realizamos el print


grades = [10,10,8,8,4,4,1,10,9,8,8,7,7,9,9,10]

in_order_grades = sorted(grades)

total_number_grades = len(grades)

if total_number_grades % 2 != 0:
    mid = in_order_grades // 2
else:
    center1= in_order_grades[total_number_grades // 2 - 1]  
    center2 = grades[total_number_grades //2 ] 
    mid = (center1 + center2) // 2

print(mid)

```

```python
"""
9. Escribe una función que compruebe si un string dado es un palíndromo. Un palíndromo es
una secuencia de caracteres que se lee igual de izquierda a derecha que de derecha a izquierda.
Por ejemplo, la función devolverá True si recibe el string “reconocer” y False si recibe el string
“python”
"""

def palindromo (string):
    return string == string[::-1]
    

palabra = "Hola"
cosa = "cola"
animal = "oso"

print(palindromo(palabra))
print(palindromo(cosa))
print(palindromo(animal))
```

```python
"""
1. Escribe una función que reciba como entrada una lista con números y devuelva como resultado
una lista con los cuadrados de los números contenidos en la lista de entrada.
"""
def cuadratica (lista):

    lista_potencia = []
    
    for recorrer in lista:
       cuadrado = recorrer * recorrer
       lista_potencia.append(cuadrado)
        
        
    return lista_potencia
        

lista = [1,2,3,4,5]

print(cuadratica(lista))
```

```python
"""
Escribe una función que reciba números como entrada y devuelva la suma de los mismos. La
función debe ser capaz de recibir una cantidad indeterminada de números. La función no
debe recibir directamente ningún objeto complejo (lista, conjunto, etc.).
"""
def suma_total (*num):
    acumulador = 0
    
    for recorrer in num:
        acumulador += recorrer
        
    return acumulador

print(suma_total(1,2,3))

    
```

```python
"""
3. Escribe una función que reciba un string como entrada y devuelva el string al revés. Ejemplo:
si el string de entrada es ‘hola’, el resultado será ‘aloh’.
"""

def reverso (string):
    reverso_jiji = string[::-1]
    return reverso_jiji

print(reverso("casa"))
```

```python
"""
Escribe una función lambda que, al igual que la función desarrollada en el ejercicio anterior,
invierta el string recibido como parámetro. Ejemplo: si el string de entrada es ‘hola’, el
resultado será ‘aloh’.
"""
#Formula:
#variable = lambda parámetro : operación

reverseishon = lambda string : string[::-1]

print(reverseishon("casa"))
```
