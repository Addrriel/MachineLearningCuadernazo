```python
# 9. Escribe un programa que obtenga la mediana de una lista de números. Recuerda que la
#mediana M de una lista de números L es el número que cumple la siguiente propiedad: la
#mitad de los números de L son superiores a M y la otra mitad son inferiores. Cuando el
#número de elementos de L es par, se puede considerar que hay dos medianas. No obstante,
#en este ejercicio consideraremos que únicamente existe una mediana.

#Crear una lista de números
#Crear la mediana
#Si es par, se considera que hay dos medianas
#Si es impar solo habrá una
#Se considera que existe una mediana para este ejercicio
#Para el par, necesitaremos sumar las dos medianas y dividir para dos

#Creamos la lista
lista_numeros = [1,2,3,4,5,6,7,7,7,7,7,87,2,3,4,234,234,2,2,34,2,3,24,234,5,15,15,12,35,16,61,3,213,123,1,21,3,1321,1,2,2,2,3,4,5]
#Necesitamos ordenar los números
lista_numeros.sort()
#Sacamos la cantidad total de elementos
cantidad_lista = len(lista_numeros)
#Luego sacaremos la mediana
mediana = cantidad_lista // 2

if cantidad_lista % 2 == 0:
    totalMediana = lista_numeros[mediana]
    print("La mediana es : ", totalMediana)
else:
    #totalMediana = (un número antes del índice de la mediana + el índice actual de la mediana) y todo eso dividido para 2
    totalMediana = (lista_numeros[mediana - 1] + lista_numeros[mediana]) / 2
    print("La mediana es : ", totalMediana)

    
    

```
