```python
#1. Escribe una función que reciba una ruta de un fichero de texto y una cadena de caracteres a buscar y determine si la cadena aparece en el fichero.
import os

def buscar_repeticiones(fichero, palabras):
    with open(fichero, 'r') as ficherito:
        for linea in ficherito:
            if palabras in linea:
                print("esta palabra existe en el fichero")
                return
        print("esta palabra no existe en el fichero")

fichero_nuevo_ex1 = os.path.join("docs", "ejercicio1.txt")
buscar_palabra = "Bedoya"

buscar_repeticiones(fichero_nuevo_ex1, buscar_palabra)
```

```python
#2. Escribe una función que reciba una lista, una ruta destino y un número n. La función debe
#crear un fichero en la ruta especificada. El contenido del fichero serán los primeros n elementos
#de la lista. La función debe controlar de manera apropiada los posibles valores de n que estén
#fuera de rango.

import os

def mostrar_contenido(lista, ruta, n):
    #Vamos hacer control de errores
    if n < 0:
        print("Error, no puede existir números negativos")
        return
        
    if n > len(lista):
        print("Supera el tamaño de la lista")
        return

    with open(ruta, 'w') as ficherote:
        for recorrer in lista[:n]:
            ficherote.write(str(recorrer) + '\n')
    print("Se ha creado el fichero, sapin")

lista_ex2 = [12,24,32,46,52,11]
ruta = os.path.join("docs", "ejercicio2.txt")

mostrar_contenido(lista_ex2, ruta, 4)
        
```

```python
#3. Escribe una función que reciba una ruta de un fichero de texto devuelva un diccionario con la
#frecuencia de aparición de cada palabra. Ejemplo: un fichero que contenga la frase ‘es mejor
# que venga que que no venga’ devolverá el siguiente diccionario: {‘es’ : 1, ‘mejor’ : 1, ‘que’ :
# 3, ‘venga’ : 2, ‘no’ : 1}. Para dividir un string en palabras puedes hacer uso del método split.
import os

def repeticion(ruta_fichero):
    repeticiones = {}

    with open(ruta_fichero, 'r') as fichero_ex3:
        for lineas in fichero_ex3:
            palabras_en_frase = lineas.split()

            for palabra_magica in palabras_en_frase:
                if palabra_magica in repeticiones:
                    repeticiones[palabra_magica] += 1
                else:
                    repeticiones[palabra_magica] = 1
    return repeticiones


rutitax = os.path.join("docs", "ejercicio3.txt")
resultadox = repeticion(rutitax)
print(resultadox)
```
