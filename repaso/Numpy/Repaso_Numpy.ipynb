```python
#20. Considera un array con forma (6, 7, 8), ¿cuál es el índice (x, y, z) del elemento número 100? (★☆☆)

"""
Usa el row major, (x, y, z) = (6, 7, 8)

Cada fila z tiene: 8 elementos
Cada bloque y tiene: 7 × 8 = 56 elementos
Cada bloque x tiene: 7 × 8 = 56 elementos

Para X debería ser:
x = 100 // (7 × 8) = 100 // 56 = 1
restante = 100 − 56 = 44

Para Y debería ser:
y = 44 // 8 = 5
restante = 44 − 40 = 4


y z sería igual a 4 z=4

entonces: (x, y, z) = (1, 5, 4)
"""
#Creamos el array como nos pude
array_ex20 = np.arange(6 * 7 * 8).reshape(6, 7, 8)

#Unravel index convierte en un índice lineal (100)
#en un índice multidimensional
#usando el orden por defecto (order='C')
indice = np.unravel_index(100, array_ex20.shape)

print(indice)

#Verificamos que sea este:
print(array_ex20[1, 5, 4])
```
