```python
#INSTRUCCION
#5. Escribe un programa que, dada una lista de strings L, un string s perteneciente a L y un string t, reemplace s por t en L.
#El programa debe mostrar la lista resultante por pantalla.

#lÓGICA
#1) Crear una lista de strings llamada "L"
#2) Crear un string llamado S y otro llamado t
#3) Creamos un bucle for para buscar la palabra correcta mediante el índice
#4) Luego vamos a crear un if para comparar el índice con el string
#5) Finalmente vamos a reemplazar mediante el índice a s por t

#Paso 1: Crear la lista de strings
lista_nombres = ["Goku", "Vegeta", "Pícolo", "Gohan"]

#Paso 2: Crear los dos strings
s = "Pícolo"
t = "Bardock"

#Paso 3: Creamos un bucle for para buscar en el rango de la loguitud dentro de la lista L
for i in range(len(lista_nombres)):
    if lista_nombres[i] == s:
        lista_nombres[i] = t
print(lista_nombres)
```

```python
#INSTRUCCIÓN

#9. Escribe un programa que obtenga la mediana de una lista de números.
#Recuerda que la mediana M de una lista de números L es el número que cumple la siguiente propiedad: la mitad de los números de L son superiores a M y la otra mitad son inferiores.
#Cuando el número de elementos de L es par, se puede considerar que hay dos medianas. No obstante, en este ejercicio consideraremos que únicamente existe una mediana

#LÓGICA
#1) Crear la lista de números
#2) Ordenar la lista 
#3) Creamos una variable que almacene la cantidad de números
#4) Hacemos una división absoluta en otra varible (mediana) 

lista_mediana = [1,2,2,2,3,3,4,5,6,7,1,2,3,5,2,3,3,3,1,45,56,6,5,4,3,5,6,76,7,5,4,3,2,1,1,1,1,2,3,4,5]
lista_mediana.sort()
cantidad_lista = len(lista_mediana)
mediana = cantidad_lista//2

if cantidad_lista % 2 == 0:
    mediana_total = lista_mediana[mediana]
else:
    mediana_total = (lista_mediana[mediana -1] + lista_mediana[mediana] / 2)

print(mediana_total)
```
