```python
#Imprime los números pares del 1 al 20.
for numeros_pares in range(1,21):
    if numeros_pares % 2 == 0:
        print(numeros_pares)
    numeros_pares+=1
```

```python
#Imprime los números impares del 1 al 20.
for numeros_impares in range(1,21):
    if numeros_impares % 2 != 0:
        print(numeros_impares)
    numeros_impares += 1
```

```python
#Cuenta cuántos números hay en la lista [1,3,5,7,9,2,4,6,8] que son pares.
lista_num2 = [1,3,5,7,9,2,4,6,8]
contador = 0

for numero in lista_num2:
    if numero % 2 == 0:
        contador += 1

print("Cantidad de números pares:", contador)
```

```python
#Pide un número al usuario e imprime su tabla de multiplicar del 1 al 10.
num_user = int(input("Coloca tu número para darte la tabla hasta el 10: "))
contador = 1

while contador <= 10:
    resultado = num_user * contador
    print(num_user, "x", contador, "=", resultado)
    contador += 1

```

```python
#Dada una lista, encuentra el número mayor sin usar max().
lista_prev = [1,2,3,4,5,6,7,8,9]

mayor = lista_prev[0]   # suponemos que el primero es el mayor

for numero in lista_prev:
    if numero > mayor:
        mayor = numero

print("El número mayor es:", mayor)
```
