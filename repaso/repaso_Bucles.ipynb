```python
#3. Escribe un programa que use un bucle while para calcular el factorial de un número n ingresado por el usuario.
factorial_user = int(input("Ingrese un número entero"))
acumulador = 1
contador = 0

while contador < factorial_user:
    acumulador+= 1
    contador *= acumulador
    print(contador)
    
```

```python
#4. Dada la cadena "abracadabra", usa un bucle for para contar cuántas veces aparece la letra "a".
lachiconmula = "abracadabra"

cadena = "abracadabra"
contador = 0

for letra in cadena:
    if letra == "a":
        contador += 1

print("La letra 'a' aparece", contador, "veces")
 
    
```

```python
#5. Usa un bucle while para encontrar el primer número mayor que 50 que sea divisible tanto por 3 como por 5.
#Creamos el número 51

numero = 51  # primer número mayor que 50 porque necesitamos todo ese rango

#Mientras este sea true
while True:
    #Haremos un if donde sea divisible para 3 y 5
    if numero % 3 == 0 and numero % 5 == 0:
        #Imprime los números que sean con esa condición
        print("El primer número mayor que 50 divisible por 3 y 5 es:", numero)
        #Cuando encuentre el primer número y el ciclo se quiera repetir, el break hace que lo cancele
        break
    #Vamos sumando uno a uno hasta que la condición se cumpla
    numero += 1
```
