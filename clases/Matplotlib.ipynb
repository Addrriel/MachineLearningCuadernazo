# Matplotlib
Es una librería estándar de representación gráfica, es de bajo nivel, es decir, fácil de usar pero difícil de dominar y puede hacer gráficas y diagramas

### Elementos de Visualización

- **Gráfico** = Tipo de representación dentro de una subfigura o figura. Existen gran cantidad de tipos de gráficos como: Barras, Líneas, Tarta, etc
- **Subfigura** = Representación dentro de una figura
- **Figura** = Marco donde se representan de manera conjunta el gráfico

### Figura y Ejes (Figure and Axes)

- **Figura (Figures):** Objeto en el que residen todos los gráficos (uno o más)
- **Ejes (Axes):** Son cada uno de los subplots(o gráficos) dentro de la Figura

```python
#1) Creamos múltiples figuras dentro de un mismo marco

#subplots(2 filas de gráficos, 3 columnas de gráficos)
figura, ejes = plt.subplots(3,2) #filas x col
print(type(ejes))
print(ejes.shape)
```

```python
#Vamos hacer un ejemplo utilizando subplots

#1) Vamos a generar los datos para las gráficas

altura_hombres = np.arange(0,20)
altura_mujeres = np.arange(-20,0)
peso_hombres = np.random.randint(100,size=(20))
peso_mujeres = np.random.randint(300,size=(20))

#2) Vamos a crear el marco y las figuras
marco, ejes = plt.subplots(1,2) #1 fila, 2 col

#3) Ahora vamos a colocar los datos según la figura
#Se coloca mediante los índices
ejes[0].plot(altura_hombres, peso_hombres)
ejes[1].plot(altura_mujeres, peso_mujeres)

#4) Ahora mostramos la gráfica
plt.show()
```

```python
#Ahora vamos a crear un nuevo eje con el eje x compartido
#Básicamente sirve para compartir 2 gráficas en un mismo eje

#1) Creamos la figura
figura, eje = plt.subplots(1,1) #una gráfica

#2) Creamos el primer valor para eje x
eje_x = np.arange(12) #tendrá 12 valores
eje.plot(eje_x, 'r') #la r es del color, es "red"

#3) Creamos el enlace para compartir 2 valores en un eje
eje2 = eje.twinx()#Con esto,creamos un eje "y" y comparte el x

#4) Creamos el segundo valor para el mismo eje
eje2.plot(eje_x ** 2, "b") #Hacemos que la gráfica sea el doble que la primera, de color azúl

plt.show()
```

```python
#Vamos a crear un ejemplo parecido a una gráfica real

#1) Creamos el marco y las figuras
marco, ejes = plt.subplots(1,2) #2 gráficas

#2) Preparamos dos gráficas
ejes[0].plot(altura_hombres, peso_hombres)
ejes[1].plot(altura_mujeres, peso_mujeres)

#3) Tanto los ejes como las figuras pueden modificarse a través de funciones
ejes[0].set_ylim(0,60)
ejes[0].set_xlim(-2,22)

ejes[0].set_xlabel("Altura")
ejes[0].set_ylabel("Peso")

ejes[0].set_ylim(0,100)
ejes[0].set_xlim(-10,2)

ejes[1].set_xlabel("Altura")
#loc es para poner la gráfica de manera central y labelpad es para ajustar el texto a una ubicación relevante
ejes[1].set_ylabel("Peso", loc="center", labelpad= -300)

#4) Mostramos la gráfica
plt.show()
```

## .plot()

Se pueden poner mútiples parámetros

Su estructura es:
`plt.plot(eje x, eje y, estilo, label, linewidth=1, alpha=1)`
- **Estilo** = Es el string donde ponemos el color como 'r', 'b', 'g', rojo azul y verde
- **Label** = Es la referencia que se tendrá para la leyenda (básicamente el nombre del gráfico)
- **Linewidth** =  Es el grosor de la línea
- **Alpha**: Es la opacidad/transparencia que tendrá la línea

Otros marcadores que podrás visualizar [marcadores](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
#Vamos hacer el primer ejercicio

#1) Creamos los datos
x_data = np.arange(10)
y_data = np.random.randint(100,size=(10))

#2) Comenzamos a estructurar nuestra gráfica

marco, ejes = plt.subplots(1,1) #1 fila, 1 col
#en los estilos puedes poner color forma y forma de las uniones
#red, triangle, rayas --
ejes.plot(x_data, y_data, 'rv--',label= "dólares",linewidth=1, alpha=0.6)

#3) escribir "best" asegura que matplotlib coloque de la mejor manera el título de la gráfica
ejes.legend(loc="best")

#4) mostrar gráfica
plt.show()
```

### Representar más de un Gráfico en una Subfigura
Básicamente quiere decir poner dos gráficos dentro de una imagen estadística

```python
#1) Creamos los datos
datos_financieros = np.arange(10)
figura, ejes = plt.subplots(1,1)

#2) Configuramos la estructura de la primera gráfica

ejes.plot(datos_financieros, np.random.randint(100,size=(10)), 'gD-.', label="Primera", linewidth= 1, alpha = 1)

#3) Ahora creamos el segundo
ejes.plot(datos_financieros, np.random.randint(100,size=(10)), 'rD-.', label = "Segunda", linewidth = 1)

#4) Ahora creamos la tercera gráfica
ejes.plot(datos_financieros, np.random.randint(50,80, size=(10)),'cD-.', label = "Tercera",linewidth=2, alpha=0.4)

#5) Ordenamos el título dentro del cuadro
ejes.legend(loc='best')

#6) Hacemos una cuadrícula (literal por los jaja's)
plt.grid()

#7) Mostramos el gráfico con plt.show
plt.show()
```

## Anotaciones

Formas de añadir información sobre el gráfico
- `ejes.text(datos_x, datos_y, 'texto que escribir')`
- `ejes.text(datos_x , datos_y , xytext, xycoords, arrowprops)`

Puedes buscar más métodos de esto [aquí](https://matplotlib.org/stable/api/axes_api.html)

```python
#Hagamos un ejercicio para ponerlo en práctica
#Cuidado con las dimensiones absolutas

#1) Creamos los datos para "x" y para "y"
datos_y = np.random.randint(100,size=(10))
datos_x = np.arange(10)

#2) Creamos la estructura del gráfico
marco, ejes = plt.subplots(1,1) #habrá un gráfico

#3) Insertamos los datos en el gráfico
ejes.plot(datos_x, datos_y)

#4) Ponemos solo el mensaje
#ejes.text(x_data[1],y_data[1],'Low income') 

#4.2) Ponemos el mensaje completo
#La estructura es xy=(valor_x[valor 1], valor_y[valor 1])
#luego xytext=(en qué punto de "x" y "y" se pondrá el mensaje)
#Luego hacemos un arrowprops=dict(facecolor = color de la flecha y shrink es el grosor de la flecha)
ejes.annotate('Menor valor', xy=(datos_x[1], datos_y[1]), xytext=(5,10), arrowprops= dict(facecolor='green', shrink = 1.0))

plt.show()
```

## Tipos de Gráficos

Ejemplos disponibles como referencia por [aquí]( https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

```python
#Ahora vamos hacer Barras Horizontales

#1) Creamos la estructura (por default es 1x1)
marco, ejes = plt.subplots()

#2) Creamos datos de ejemplo
cargo = ('PhD', 'MSc', 'BSc', 'N/A')
posicion_y = np.arange(len(cargo))
personas_cargo = [5,45,25,25]

#3) Como so barras horizontales, nos enfocaremos en el eje Y
ejes.barh(posicion_y, personas_cargo, align= 'center', color='blue', edgecolor = 'red')

#4) Definimos las posiciones que tendrá Y
ejes.set_yticks(posicion_y)

#5) Colocamos las etiquetas de los cargos para las barras
ejes.set_yticklabels(cargo)

#6) Invertimos los datos (se leen al revés de como deberían ser)
#ejes.invert_yaxis()

#7) Métemos una etiqueta para la parte horizontal
ejes.set_xlabel('Percentage of workforce')

#8) Título del marco 
ejes.set_title('Nivel de Educación en la Compañía')


plt.show()
```

```python
# Ahora vamos hacer con un ejemplo más completo

#1) Creamos los datos

ls_count = (284487, 244560, 208493, 196764)
cities = ('Madrid', 'Barcelona', 'Sevilla', 'Valencia')
data = np.array(ls_count)/1000
y_pos = np.arange(data.size)


#2) Creamos valores default para la gráfica (por el momento solo son variables xd)
width = 0.35 #este lo usaremos para el ancho de las barras
opacity = 0.7 #este lo usaremos para el alpha

#3) Creamos la estructura de la gráfica
figura, ejes = plt.subplots()
rects1 = ejes.barh(y_pos, data, width, alpha=opacity) #aquí se aplican las variables

#4) Vamos a poner el color en cada una de las barras
rects1[0].set_color('#6ade30')
rects1[1].set_color('#d6ee39')
rects1[2].set_color('#ffcb59')
rects1[3].set_color('#872f29')

#5) Vamos a definir bien las posiciones de "y" y vamos a colocar una etiqueta en "y" que es para ciudades
ejes.set_yticks(y_pos)
ejes.set_yticklabels(cities)

#6) Vamos añadir etiquetas para hacer referencia a los campos de "x" y "y"
ejes.set_ylabel('Ciudades')
ejes.set_xlabel('Km de carril bici')

#7) Vamos a crear la leyenda de todas las ciudades
#estructura ejes.legend((grafica[0], grafica[1], etc), nombre_ciudades)

ejes.legend((rects1[0], rects1[1],rects1[2],rects1[3]), cities)

plt.show()
```

```python
# Ahora vamos hacer el stack plot

#1) Creamos los datos

x = [2008, 2009, 2010, 2011, 2012]
iOS = [35, 25, 20, 17, 18]
Android = [0, 15, 35, 45, 60]
WindowsPhone = [1, 3, 5, 5, 2]
Others = [64,57,40,33,20]

#2) Esto serviría para poner a todos dentro de una matriz
#cada dato sería una fila
y = np.vstack([iOS, Android, WindowsPhone,Others])
print(y)

print("------------------------------------------")

#3) Creamos etiquetas para cada uno
labels = ["iOS ", "Android", "WindowsPhone", "Others"]

#4) Creamos la estructura para la gráfica
figura, ejes = plt.subplots()

#5) Creamos un stack para que dibuje dibuja áreas rellenas
#una encima de otra para mostrar cómo contribuyen varias
#series al total a lo largo del eje X

#estructura: (x, cada uno de los datos, labels=labels creados en esa lista)
ejes.stackplot(x, iOS, Android, WindowsPhone, Others,labels=labels)

# 6) Esto es para aplicarlo en "x" y "y"
#ejes.stackplot(x, y, labels=labels)

#7) Definimos las posiciones de x
ejes.set_xticks(x)

#8) Creamos la leyenda que dirá que esté arriba a la izquierda
ejes.legend(loc='upper left')

#9) Mostrar la gráfica
plt.show()
```

```python
#Ahora vamos hacer el Scatter plot

from numpy.random import rand

#1) Creamos la esctructura de la gráfica
figura, ejes = plt.subplots()

#2) Creamos una variable para indicar cuántos puntos aleatorios
#se van a generar y dibujar por cada color
n = 700

#3) Creamos un for para recorrer la lista de color y
#en cada vuelta del bucle, asigna un color distinto a la variable

for color in ["red","green","blue"]:
    #genera n números aleatorios para x = 2 y "y" = n
    x,y = rand(2,n)

    #generamos la escala
    scale = 25.0 * rand(n) #Cada punto tendrá un tamaño diferente

    ejes.scatter(x, y, c=color, s=scale, label=color, alpha=0.3, edgecolors='none')

ejes.legend()
plt.show()
```

## Representación Gráfica en Pandas
Pandas ofrece interfaz para poder dibujar con Series o DataFrames, se puede utilizar matplotlib internamente y es fácil de utilizar

```python
#Primero vamos a crear una matriz 2D 10x3
matriz_random = np.random.randint(100,size=(10,3)) #10 filas x 3 col
#print(matriz_random)

#2) Vamos a crear un data frame de la matriz, y las col se llamaran A, X, C
frame = pd.DataFrame (matriz_random, columns= list('AXC'))
#3) Creamos una gráfica default
frame.plot()
#4) Mostramos la gráfica
plt.show()
#5) Mostramos el dataframe (solo se ven los datos gracias al pandas)
display(frame)
```

## frame.plot()
la estructura es:
`frame.plot(valor_x, valor_y, label=<Series>, ax,style,kind,xticks,yticks,title,subplots)`



- (Solo para Series) **label** = referencia para la leyenda
- **ax** = subfigura para dibujar los datos
- **style** = estilo de la línea
- **kind**: tipo de gráfico (barra, pastel, area, densidad, kde, etc)
- **xticks and yticks**= valores en los ejes "X" y "Y"
- **title**: string como título
- (Solo para DataFrame) **subplots**: boolean si se desean subfiguras separadas para cada columna
- (Solo para DataFrame) **"x" y "y"** se pueden usar para seleccionar columnas para cada eje

Puedes visualizar la lista de argumentos [aqui](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html)

```python
#Vamos a crear nuestro primer ejercicio de esto

#1) Creamos una matriz random de 30 números menores a 100, 10 filas x 3 col
matriz_random = np.random.randint(100,size=(10,3))

#2) Creamos el Dataframe para poder verlo mejor
frame = pd.DataFrame(matriz_random, columns=list("ABC"))

#3) Creamos la estructura de la gráfica
#el xticks hace que el rango de la gráfica sea de -2 a 11
frame.plot(style=['r-','g-','b-'], xticks= range(-2,12), title= "Líneas random", subplots = True)

#4) Colocamos la leyenda en la primera fila (arriba a la izquierda)
plt.show()
display(frame)
```

## Relaciones entre Variables

**`sns.replot()`**

```python
#1) Cargamos el dataset ya generado anteriormente
tips = sns.load_dataset("tips") #este tips no es una copia ni nada relacionado, es del propio sns
#2) generamos 5 datos al azar
tips.sample(5)
```

```python
#Ahora vamos a graficar con SNS
# variable x, variable y, tercer valor, datos, tipo de gráfico
sns.relplot(x='total_bill', y='tip', hue='sex', data=tips, kind='scatter')
#Mostramos el gráfico
plt.show()
```

```python
#Ahora vamos a
#Representar la media y la variación de una variable en función de
#otra

#Traemos el dataset a la acción
data = sns.load_dataset('flights')
#Creamos la gráfica = variable x, variable y, tipo de gráfica, datos
sns.relplot(x="year", y="passengers", kind="line", data=data)
#mostramos la gráfica
plt.show()
```

```python
#Podemos relacionar hasta 4 variables de manera automática
#Se crean dos variables a comparar: total_bill y tip
#Luego divide entre fumadores y no fumadores (smoker)
#Al final se creará una tabla (columna) por cada tipo de sex.
#Como hay 2, por eso son dos tablas
#El ancho se ajusta automáticamente con el height
sns.relplot(x= "total_bill", y="tip", hue= "smoker", col = "sex", height = 4, kind= "scatter", data=tips )
```

### Distribución de un DataSet

```python
#scatter de clases, coloreando segun clase en matplotlib
iris = sns.load_dataset('iris')#cargamos el dataset de iris
```

```python
#1) Preparamos los datos
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]

#2) Hacemos un for para que busque por clase definida
#recorre toda la lista de groups
for i in groups:
    #iris['species'] == i crea una máscara para que se quede solo con las filas de esa especie
    #[columns] ignora las demás columnas 
    #.values convierte el resultado en un array
    data.append(iris[iris['species'] == i] [columns].values)

#3) Creación de la gráfica
figura = plt.figure() #creamos el cuadro (figura vacía)
eje = figura.add_subplot() #añadimos un eje dentro de la figura

#Devuelve dos cosas en cada vuelta:
#count = índice (0, 1, 2)
#d → cada grupo de datos (array numpy)
for count, d in enumerate(data):
    #d[:,0] = Todas las filas, Columna 0 que sería "petal_length"
    #d[:,1] = lo mismo que arriba pero Columna 1, esto sería "petal_width"
    x, y = d[:,0], d[:,1]
    #recorre desde el último color en adelante
    #no tiene borde de color
    #s = tamaño de cada punto
    #label=groups[count - 1] = recorre los nombres del último para adelante
    eje.scatter(x,y,alpha=0.4,c= colors[count -1], edgecolors = 'none', s= 30, label = groups[count - 1])

#4) Ponemos las etiquetas
eje.set_xlabel(columns[0]) #Etiqueta para longuitud de pétalos
eje.set_ylabel(columns[1])# Etiqueta para anchura de pétalos
plt.legend(loc='best') #Coloca la leyenda donde mejor le parezca
plt.title('Matplot scatter plot') #Ponemos título a la gráfica
plt.show() #Mostramos la gráfica
    
```

### matplotlib
- crear una lista de años de 2000 a 2020 (eje x)
- generar datos aleatorios para el eje y (hacerlo 4 veces distintas: y0, y1, y2 e y3)
- Representar las 4 secuencias de datos aleatorios en una sola figura utilizando matplotlib
- Añade una leyenda para poder identificar cada secuencia en el gráfico
- Nombra las secuencias de la siguiente forma: “Hawaii”, “San Marino”, “Islas Feroe”,
“Guayana”
- Representar los mismos datos en 4 subfiguras (axes) distintas, parte de la misma figura
- Añade la siguiente info:
   - Título: “Datos aleatorios”
   - Leyenda
   - Nombre del eje x: “Año”
   - Nombre del eje y: “GDP”
- Añade al menos una anotación (texto y flecha) a uno de los gráficos


```python
import numpy as np
import matplotlib.pyplot as plt

# 1) crear una lista de años de 2000 a 2020 (eje x)
anios = np.arange(2000, 2021)

# 2) generar datos aleatorios para el eje
#y (hacerlo 4 veces distintas: y0, y1, y2 e y3)
y0 = np.random.randint(100, size=(21))
y1 = np.random.randint(100, size=(21))
y2 = np.random.randint(100, size=(21))
y3 = np.random.randint(100, size=(21))

#3) Representar las 4 secuencias de datos aleatorios
#en una sola figura utilizando matplotlib
plt.figure()

plt.plot(anios, y0, label="Hawaii")
plt.plot(anios, y1, label="San Marino")
plt.plot(anios, y2, label="Islas Feroe")
plt.plot(anios, y3, label="Guayana")

plt.title("Datos aleatorios")
plt.xlabel("Año")
plt.ylabel("GDP")

#4) Añade una leyenda para poder identificar cada secuencia en el gráfico
plt.legend()

# Anotación
plt.annotate(
    "Pico alto",
    xy=(2010, y0[10]),
    xytext=(2003, 90),
    arrowprops=dict(arrowstyle="->")
)

plt.show()

#Representar los mismos datos en 4 subfiguras (axes) distintas
#parte de la misma figura

fig, axes = plt.subplots(2, 2, figsize=(10, 6))
fig.suptitle("Datos aleatorios")

# Hawaii
axes[0, 0].plot(anios, y0, label="Hawaii")
axes[0, 0].set_title("Hawaii")
axes[0, 0].set_xlabel("Año")
axes[0, 0].set_ylabel("GDP")
axes[0, 0].legend()

# San Marino
axes[0, 1].plot(anios, y1, label="San Marino")
axes[0, 1].set_title("San Marino")
axes[0, 1].set_xlabel("Año")
axes[0, 1].set_ylabel("GDP")
axes[0, 1].legend()

# Islas Feroe
axes[1, 0].plot(anios, y2, label="Islas Feroe")
axes[1, 0].set_title("Islas Feroe")
axes[1, 0].set_xlabel("Año")
axes[1, 0].set_ylabel("GDP")
axes[1, 0].legend()

# Guayana
axes[1, 1].plot(anios, y3, label="Guayana")
axes[1, 1].set_title("Guayana")
axes[1, 1].set_xlabel("Año")
axes[1, 1].set_ylabel("GDP")
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

## Seaborn Ejercicio

- Carga los datos de un dataset de seaborn (distinto a ‘tips’, ‘iris’ o ‘flights’) https://github.com/mwaskom/seaborn-data
- Si para alguno de los ejercicios no encuentras ejemplos en el dataset, busca otro para completarlo
- Recuerda utilizar la versión ‘Raw’ si se utiliza la URL en el método read_csv
- Representa visualmente la relación de cada pareja de atributos (variables)
- Escoge una pareja (ambos valores contínuos y numéricos) y demuestra su correlación
- Representa los valores de dos variables (x e y, numéricas) en base a otra variable (categórica)
- Crea los gráficos que demuestren lo siguiente en el dataset seleccionado:
  - La distribución de valores (media, distribución, outliers) de una variable numérica contínua
  - Compara la distribución de esa variable con otra (contínua y numérica) para representar cuantitativamente la densidad de elementos por valor en ambas variables
  - Muestra un ejemplo en el dataset de una variable a la que un modelo de regresión lineal de orden > 1 sea más ajustado que uno de orden == 1


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#1) Cargamos un dataset distinto a los anteriores
cuakcuak = sns.load_dataset("penguins")

#2) Mostramos las primeras filas
print(cuakcuak.head())

#3) Representa visualmente la
#relación de cada pareja de atributos (variables)
sns.pairplot(cuakcuak, hue="species")
plt.show()

#4) Escoge una pareja (ambos valores contínuos y numéricos)
#y demuestra su correlación
sns.regplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data=cuakcuak
)
plt.title("Correlación entre longitud y profundidad del pico")
plt.show()

#5) Representa los valores de dos variables (x e y, numéricas)
#en base a otra variable (categórica)
sns.scatterplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species", #aquí va la categoría
    data=cuakcuak
)
plt.title("Longitud vs Profundidad del pico por especie")
plt.show()

#6) La distribución de valores (media, distribución, outliers)
#de una variable numérica contínua
sns.boxplot(
    y="body_mass_g",
    data=cuakcuak
)
plt.title("Distribución del peso corporal")
plt.show()

#7) Compara la distribución de esa variable
#con otra (contínua y numérica) para representar cuantitativamente
#la densidad de elementos por valor en ambas variables
sns.kdeplot(
    cuakcuak["bill_length_mm"],
    label="Bill Length",
    fill=True
)

sns.kdeplot(
    cuakcuak["bill_depth_mm"],
    label="Bill Depth",
    fill=True
)

plt.title("Comparación de densidad entre variables")
plt.legend()
plt.show()

#8) Muestra un ejemplo en el dataset de una variable a la que
#un modelo de regresión lineal de orden > 1 
#sea más ajustado que uno de orden == 1
sns.regplot(
    x="bill_length_mm",
    y="body_mass_g",
    data=cuakcuak,
    order=2
)
plt.title("Regresión polinómica (orden 2)")
plt.show()
```
