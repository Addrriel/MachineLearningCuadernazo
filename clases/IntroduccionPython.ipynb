<p style="max-width:900px; margin:0 auto 12px auto; font-family:'Segoe UI', Tahoma, sans-serif; font-size:16px; line-height:1.7; color:#444;">

<p style="max-width:900px; margin:0 auto 12px auto; font-family:'Segoe UI', Tahoma, sans-serif; font-size:16px; line-height:1.7; color:#444;">

    <code style="background:#f1f1f1; padding:4px 8px; border-radius:6px; color:#c7254e;">suma: "5" + 3</code>,

<p style="max-width:900px; margin:0 auto 12px auto; font-family:'Segoe UI', Tahoma, sans-serif; font-size:16px; line-height:1.7; color:#444;">

<h3 style="text-align:center; font-family:'Segoe UI', Tahoma, sans-serif; font-size:24px; color:#2c3e50; margin-top:40px;">
    REPL (READ, EVALUATE, PRINT AND LOOP)
</h3>

<p style="max-width:900px; margin:0 auto 15px auto; font-family:'Segoe UI', Tahoma, sans-serif; color:#444; line-height:1.7;">
    Es un sistema interactivo para comunicarse con un lenguaje (Python en este caso) directamente con el ordenador.
</p>

<ul style="max-width:850px; margin:0 auto 25px auto; font-family:'Segoe UI', Tahoma, sans-serif; color:#444; line-height:1.6;">
    <li><strong>READ</strong> – Permite al ordenador leer la entrada.</li>
    <li><strong>EVALUATE</strong> – El código es procesado.</li>
    <li><strong>PRINT</strong> – Muestra los resultados en pantalla.</li>
    <li><strong>LOOP</strong> – Continúa con la conversación.</li>
</ul>

<h3 style="text-align:center; font-family:'Segoe UI', Tahoma, sans-serif; font-size:24px; color:#2c3e50;">
    Ficheros
</h3>

<p style="max-width:900px; margin:0 auto 12px auto; font-family:'Segoe UI', Tahoma, sans-serif; color:#444; line-height:1.7;">
    Permiten guardar nuestros programas para poder ejecutarlos después. Generalmente son archivos de texto con extensión <strong>.py</strong>. En Linux es importante incluir al inicio del archivo el
    <code style="background:#f1f1f1; padding:4px 8px; border-radius:6px;">#!/usr/bin/env python3</code>.
</p>

<p style="max-width:900px; margin:0 auto 12px auto; font-family:'Segoe UI', Tahoma, sans-serif; color:#444; line-height:1.7;">
    Un archivo puede llamarse <strong>script</strong> o <strong>programa</strong> cuando es el principal, y <strong>módulo</strong> cuando se importa usando <code>import</code>.
</p>

<p style="max-width:900px; margin:0 auto 12px auto; font-family:'Segoe UI', Tahoma, sans-serif; color:#444; line-height:1.7;">

    <br>
    <code style="background:#f1f1f1; padding:6px 10px; border-radius:6px;">python hola_soy_alejandro.py</code>
</p>

<p style="max-width:900px; margin:0 auto 15px auto; font-family:'Segoe UI', Tahoma, sans-serif; color:#444; line-height:1.7;">

</p>

    Performance Sample, NumPy vs Plain Python
</h3>

```python
import numpy as np #Importamos la librería de numpy para poder medir por los arrays
my_arr = np.arange(10000000) #Utilizamos Numpy Arrays

my_list = list(range(10000000)) #Utilizamos listas

#es un comando de Jupyter para medir cuánto tarda el bloque.
#Ejecuta 10 veces la operación
#NumPy multiplica todos los valores de una sola vez usando código optimizado en C.
%time for _ in range(10): my_arr2 = my_arr * 2 

#También se ejecuta 10 veces.
#Aquí Python tiene que recorrer elemento por elemento con un for.
#Esto es mucho más lento porque cada operación se hace en Python puro.
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]
```
