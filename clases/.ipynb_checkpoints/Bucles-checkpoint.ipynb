```python
#Bucles

#Es un mecanimos que nos sirve para iterar, nos sirve para repetir algo



#Bucle While
#Sintaxis
#while condicion:
    #ejecución random
primera_condicion = 0

#Se hace bucle infinito porque decimos
# Mientras primera condición sea menor a 10 entonces imprimirá la condicion
while primera_condicion < 10:
    print(primera_condicion)
```

```python
#¿Cómo arreglamos eso?
primera_condicion = 0

while primera_condicion < 10:
    print(primera_condicion)
    primera_condicion += 2

#Así va sumando de 2 en 2 hasta llegar al 10, una vez pase esa condición, el bucle termina
print("Se detuvo el bucle y si ves este texto es que ya salió del bucle")
```

```python
#otra forma de realizarlo es la siguiente
primera_condicion = 0

while primera_condicion < 10:
    print(primera_condicion)
    primera_condicion += 2
#En python, se puede meter un else a este ciclo xD
else: 
    print("mi condición ahora es igual o mayor a 10")

print("Se detuvo el bucle y si ves este texto es que ya salió del bucle")
```

```python
#Podemos meter condicionales y bucles

segunda_condicion = 0

while segunda_condicion < 10:
    print(segunda_condicion)
    segunda_condicion += 2
    if segunda_condicion == 10:
        print("El bucle ya alcanzo el 10")
    else:
        print("Aún no alcanza el 10")
print("Se detuvo el bucle y si ves este texto es que ya salió del bucle")
```
